```python
import pandas as pd
import wbdata
```

```python
# indicadores para descargar
indicadores = {
    "NY.GNP.PCAP.CD": "GNI per capita, Atlas method (current US$)",
    "SP.DYN.TFRT.IN": "Fertility rate, total (births per woman)",
    "SP.DYN.LE00.IN": "Life expectancy at birth, total (years)",
    "SH.DYN.MORT": "Mortality rate, under-5 (per 1,000 live births)",
    "SL.UEM.TOTL.ZS": "Unemployment, total (% of total labor force) (modeled ILO estimate)"
}

# códigos de país de América Latina y el Caribe
paises_latam_caribe = ['ATG', 'ARG', 'BHS', 'BRB', 'BLZ', 'BOL', 'BRA', 'CHL', 'COL',
                       'CRI', 'CUB', 'DMA', 'DOM', 'ECU', 'SLV', 'GRD', 'GTM', 'GUY',
                       'HTI', 'HND', 'JAM', 'MEX', 'NIC', 'PAN', 'PRY', 'PER', 'PRI',
                       'KNA', 'LCA', 'VCT', 'SUR', 'TTO', 'URY', 'VEN']

# Descargar los datos de los indicadores para los países de América Latina y el Caribe
data = wbdata.get_dataframe(indicadores, country=paises_latam_caribe, convert_date=True)


data = data.reset_index()


data["date"] = data["date"].dt.year

df = data

while df[df['date']==df['date'].max()].isnull().sum(axis=1).sum()/len(df[df['date']==df['date'].max()]) > 2:
    df = df[df['date'] != df['date'].max()]

df = df.groupby('country').apply(lambda x: x.nlargest(30, 'date')).reset_index(drop=True)
```

Aumentar el GNI per capita en al menos un 10% en razón del GNI per capita a nivel regional alcanzado para el 2021 en comparación con el 2016.

```python
df_GNI = df[['date','GNI per capita, Atlas method (current US$)']].groupby('date').mean()
df_GNI['percentage change'] = df_GNI['GNI per capita, Atlas method (current US$)'].pct_change(5)*100
df_GNI.tail(10)
```

Reducir la tasa de fertilidad en al menos un 8% en razón de la tasa de fertilidad a nivel regional alcanzada para el 2021 en comparación con el 2016.

```python
df_Fertility = df[['date','Fertility rate, total (births per woman)']].groupby('date').mean()
df_Fertility['percentage change'] = df_Fertility['Fertility rate, total (births per woman)'].pct_change(5)*100
df_Fertility.tail(7)
```

Aumentar la esperanza de vida en al menos un 3%% en razón de la esperanza de vida a nivel regional alcanzada para el 2021 en comparación con el 2011.

```python
df_Life = df[['date','Life expectancy at birth, total (years)']].groupby('date').mean()
df_Life['percentage change'] = df_Life['Life expectancy at birth, total (years)'].pct_change(10)*100
df_Life.tail(11)
```

Reducir la tasa de mortalidad infantil en al menos un 10% en razón de la tasa de mortalidad infantil a nivel regional alcanzada para el 2021 en comparación con el 2016.

```python
df_Mortality = df[['date','Mortality rate, under-5 (per 1,000 live births)']].groupby('date').mean()
df_Mortality['percentage change'] = df_Mortality['Mortality rate, under-5 (per 1,000 live births)'].pct_change(5)*100
df_Mortality.tail(7)
```

Reducir el porcentaje de desempleo en al menos un 10% en razón del porcentaje de desempleo a nivel regional alcanzado para el 2021 en comparación con el 2020.

```python
df_Unemployment = df[['date','Unemployment, total (% of total labor force) (modeled ILO estimate)']].groupby('date').mean()
df_Unemployment['percentage change'] = df_Unemployment['Unemployment, total (% of total labor force) (modeled ILO estimate)'].pct_change(1)*100
df_Unemployment.tail(7)
```
